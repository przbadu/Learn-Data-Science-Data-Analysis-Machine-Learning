## Object Oriented way

**First method**

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

**Second method**

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
fig.show()
```

**Third method**

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

## Matplotlib example workflow

1. Import matplotlib and get it ready for plotting in Jupyter
2. Prepare data
3. Setup plot
4. Plot the data
5. Customize plot
6. Save & show

```python
# import
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4]
y = [11, 22, 25, 44]

# setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # figsize=(width, height)

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save and show
fig.savefig("assets/sample-plot.png")
# plt.imread('assets/sample-plot.png')
```

### Make a plot from dictionary

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Price ($)");
```

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))

# customize labels and title
ax.set_xlabel("Price ($)")
ax.set_ylabel("Nut Butter")
ax.set_title('Prices of Nut Butters')
plt.show()
```

### Histogram

```python
x = np.random.randn(100)

fig, ax = plt.subplots()
ax.hist(x);
```

#### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set_xlabel("Price ($)")
ax3.set_ylabel("Nut Butter")
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(subplots=True);
```

```python
### Improve the visualization of those sub-plots
num_columns = len(heart_disease.columns)
num_rows = (num_columns + 3) // 4 # Calculate the numer of rows needed

# Create a figure with a specified size
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4 * num_rows))
fig.tight_layout(pad=3.0)

# Flatten axes for easy iteration
axes = axes.flatten()

for i, col in enumerate(heart_disease.columns):
  heart_disease[col].plot.hist(ax=axes[i], title=col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])


plt.show();
```

## Visualize using OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# Object-Oriented method using pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize the plot
ax.set(title='Heart Disease and Chlosterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--')
```

```python
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['thalach'])
ax.set(title='Heart Disease and Chlosterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(), title='thalach')
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
# Subplot of chol, age, thalache
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel='Age',
        ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
ax1.set(title="Heart Disease and Cholesterol Levels",
        xlabel='Age',
        ylabel='Max heart rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
import matplotlib.pyplot as plt

# Example DataFrame (replace with your actual DataFrame)
# Assuming 'over_50' is already defined with 'age', 'chol', 'thalach', and 'target' columns

# Set a valid style for the plot
# ['Solarize_Light2', '_classic_test_patch', '_mpl-gallery', '_mpl-gallery-nogrid',
# 'bmh', 'classic', 'dark_background', 'fast', 'fivethirtyeight', 'ggplot', 'grayscale',
# 'seaborn-v0_8', 'seaborn-v0_8-bright', 'seaborn-v0_8-colorblind', 'seaborn-v0_8-dark',
# 'seaborn-v0_8-dark-palette', 'seaborn-v0_8-darkgrid', 'seaborn-v0_8-deep',
# 'seaborn-v0_8-muted', 'seaborn-v0_8-notebook', 'seaborn-v0_8-paper', 'seaborn-v0_8-pastel',
# 'seaborn-v0_8-poster', 'seaborn-v0_8-talk', 'seaborn-v0_8-ticks', 'seaborn-v0_8-white',
# 'seaborn-v0_8-whitegrid', 'tableau-colorblind10']
plt.style.use('seaborn-v0_8-paper')

# Create subplots
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Define a colormap
cmap = plt.get_cmap('viridis')

# Add data to ax0 with customized colors
scatter0 = ax0.scatter(x=over_50['age'],
                       y=over_50['chol'],
                       c=over_50['target'],
                       cmap=cmap,
                       edgecolor='k',  # Add black edge color to markers
                       alpha=0.7)      # Make markers slightly transparent
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel='Age',
        ylabel='Cholesterol')
ax0.legend(*scatter0.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--',
            color='red')               # Customize the color of the horizontal line
ax0.axvline(x=over_50['age'].mean(),
            linestyle='--',
            color='blue')              # Add a vertical line at the mean age for additional information

# Add data to ax1 with customized colors
scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap=cmap,
                       edgecolor='k',  # Add black edge color to markers
                       alpha=0.7)      # Make markers slightly transparent
ax1.set(title="Heart Disease and Maximum Heart Rate",
        xlabel='Age',
        ylabel='Max Heart Rate')
ax1.legend(*scatter1.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--',
            color='red')               # Customize the color of the horizontal line
ax1.axvline(x=over_50['age'].mean(),
            linestyle='--',
            color='blue')              # Add a vertical line at the mean age for additional information

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

# Improve layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```
