## Object Oriented way

**First method**

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

**Second method**

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
fig.show()
```

**Third method**

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

## Matplotlib example workflow

1. Import matplotlib and get it ready for plotting in Jupyter
2. Prepare data
3. Setup plot
4. Plot the data
5. Customize plot
6. Save & show

```python
# import
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4]
y = [11, 22, 25, 44]

# setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # figsize=(width, height)

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save and show
fig.savefig("assets/sample-plot.png")
# plt.imread('assets/sample-plot.png')
```

### Make a plot from dictionary

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Price ($)");
```

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))

# customize labels and title
ax.set_xlabel("Price ($)")
ax.set_ylabel("Nut Butter")
ax.set_title('Prices of Nut Butters')
plt.show()
```

### Histogram

```python
x = np.random.randn(100)

fig, ax = plt.subplots()
ax.hist(x);
```

#### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set_xlabel("Price ($)")
ax3.set_ylabel("Nut Butter")
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

```python
car_sales = pd.read_csv("data/car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales['Price'].str.replace(r'[\$\,\.]', '', regex=True)
car_sales['Price'] = car_sales['Price'].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x='Sales Date', y='Total Sales');
```

```python
car_sales['Price'] = car_sales['Price'].astype(int)
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

ax1.plot(car_sales['Sales Date'], car_sales['Total Sales'])
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax3.hist(car_sales['Total Sales'])
ax4.hist(car_sales['Make'])
```

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(subplots=True);
```

```python
### Improve the visualization of those sub-plots
num_columns = len(heart_disease.columns)
num_rows = (num_columns + 3) // 4 # Calculate the numer of rows needed

# Create a figure with a specified size
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4 * num_rows))
fig.tight_layout(pad=3.0)

# Flatten axes for easy iteration
axes = axes.flatten()

for i, col in enumerate(heart_disease.columns):
  heart_disease[col].plot.hist(ax=axes[i], title=col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])


plt.show();
```
